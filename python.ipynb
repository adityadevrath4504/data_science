```python
import json
```

```python
# lets write a function to load the data
def load_data(filename):
    with open (filename,"r") as f:
       data=json.load(f)
    return data
    
```

```python
data=load_data("data.json")
```

```python
# write a function to display user and their connection
def display_users(data):
    print("users and their connections\n")
    for user in data['users']:
        print(f"ID:{user['id']}.{user['name']} is friend with:{user['friends']} and like pages are {user['liked_pages']}")
    print("\npages infromation")
    for page in data['pages']:
        print(f"{page['id']}:page{page['name']}")
display_users(data)
    
```
